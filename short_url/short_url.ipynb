```python
import hashlib
import tkinter as tk
from tkinter import messagebox

class UrlShortener:
    def __init__(self):
        self.url_mapping = {}
        self.base_url = "http://short.url/"

    def shorten_url(self, original_url):
        hash_value = hashlib.md5(original_url.encode()).hexdigest()[:6]
        short_url = self.base_url + hash_value
        self.url_mapping[short_url] = original_url
        return short_url

    def expand_url(self, short_url):
        original_url = self.url_mapping.get(short_url)
        return original_url

def shorten_url():
    original_url = entry_original.get()
    if original_url:
        short_url = url_shortener.shorten_url(original_url)
        entry_short.delete(0, tk.END)
        entry_short.insert(0, short_url)
    else:
        messagebox.showerror("Error", "Please enter a valid URL")

def expand_url():
    short_url = entry_short.get()
    if short_url:
        original_url = url_shortener.expand_url(short_url)
        if original_url:
            messagebox.showinfo("Original URL", f"Original URL: {original_url}")
        else:
            messagebox.showerror("Error", "Short URL not found")
    else:
        messagebox.showerror("Error", "Please enter a valid short URL")


url_shortener = UrlShortener()

root = tk.Tk()
root.title("URL Shortener")


label_original = tk.Label(root, text="Enter Original URL:")
label_original.grid(row=0, column=0, padx=10, pady=10)

entry_original = tk.Entry(root, width=40)
entry_original.grid(row=0, column=1, padx=10, pady=10)

label_short = tk.Label(root, text="Shortened URL:")
label_short.grid(row=1, column=0, padx=10, pady=10)

entry_short = tk.Entry(root, width=40)
entry_short.grid(row=1, column=1, padx=10, pady=10)

button_shorten = tk.Button(root, text="Shorten URL", command=shorten_url)
button_shorten.grid(row=2, column=0, columnspan=2, padx=10, pady=10)

button_expand = tk.Button(root, text="Expand URL", command=expand_url)
button_expand.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

# Start the GUI loop
root.mainloop()
```
